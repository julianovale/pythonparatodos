```
import matplotlib.pyplot as plt
```

```
class Ponto:
  def __init__(self, x = 1.0, y = 1.0, symbol = 'o', color = 'blue'):
    self.x = x
    self.y = y
    self.symbol = symbol
    self.color = color
    self.fig = plt.figure()
    self.ax = plt.axes()

  def desenha(self):
    self.ax.plot(self.x, self.y, self.symbol, color = self.color)
    self.ax.grid()
    self.fig.show()
```

```
p = Ponto()
p.desenha()
```

```
g = Ponto(4, 8, "*", 'red')
g.desenha()
```

```
class Circulo(Ponto):
  def __init__(self, x = 1.0, y = 1.0, symbol = '.', ccolor = 'red', raio = 0.25):
    super().__init__(x, y, symbol)
    self.raio = raio
    self.ccolor = ccolor

  def desenha(self):
    draw_circle = plt.Circle((self.x, self.y), self.raio, 
                             fill = True, color=self.ccolor)
    self.fig.gca().add_artist(draw_circle)
    x = self.x
    y = self.y
    r = self.raio
    xmin = x - 1.5*r
    xmax = x + 1.5*r
    ymin = y - 1.5*r
    ymax = y + 1.5*r
    self.ax.set(xlim=(xmin, xmax),ylim=(ymin, ymax))
    self.ax.grid()
    self.fig.show()
```

```
c = Circulo(ccolor = 'yellow')
c.desenha()
```

```
from random import seed
from random import randint
```

```
seed(1)

L = []
cdic = {1: "red", 2: "blue", 3: "yellow", 4: "green"}

for i in range(1,5):
  xc = randint(0,10)
  yc = randint(0,10)
  L.append(Ponto(x = xc, y = yc, color=cdic[i]))
  L.append(Circulo(x = xc, y = yc, ccolor=cdic[i]))

for g in L:
  g.desenha() # mÃ©todo polimÃ³rfico: o cÃ³digo "decide" se deve usar o desenha de Ponto ou de Circulo dependendo do caso... Genial!!!
```
