```
import pandas as pd
import io
import requests
```

```
url = 'https://github.com/julianovale/pythonparatodos/blob/main/municipios.csv'
s = requests.get(url).content
file_name = io.StringIO(s.decode('utf-8'))
d = pd.read_html(file_name)
df = d[0]
df
```

```
df.dropna(axis=1, inplace=True)
df
```
