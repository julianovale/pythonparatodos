```
import matplotlib.pyplot as plt
```

```
x = [4,3,2,1,5]
y = [10,7,8,4,3]
plt.plot(x,y,"b--")
plt.show()
```

```
x = [4,3,2,1,5]
y = [10,7,8,4,3]
plt.plot(x,y,"m--")
plt.show()
```

```
x = [4,3,2,1,5]
y = [10,7,8,4,3]
plt.plot(x,y,"b-.")
plt.scatter(x,y,marker='*',color='red')
plt.show()
```
