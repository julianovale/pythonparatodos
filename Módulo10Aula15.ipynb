```
import pandas as pd
import io
import requests
```

```
url = 'https://github.com/julianovale/pythonparatodos/blob/main/municipios.csv'
s = requests.get(url).content
file_name = io.StringIO(s.decode('utf-8'))
d = pd.read_html(file_name)
df = d[0]
df
```

```
df.dropna(axis=1, inplace=True)
df
```

```
print(df.columns.ravel())
```

```
df3 = df.loc[df['codigo_uf'] == 35, ['codigo_ibge','nome','latitude','longitude']]
df3
```

```
url_metricas = 'https://raw.githubusercontent.com/julianovale/pythonparatodos/main/ISDEL_Cidades_Estado_SPIBGE.csv'
pd_metricas = pd.read_csv(url_metricas, encoding = 'latin-1', sep = ';', decimal = ',')
pd_metricas
```

```
df = pd_metricas.dropna()
df
```

```
print(df.columns.ravel())
```

```
dfinal = df3.merge(df,left_on='codigo_ibge', right_on='Cód_Municipio')
dfinal
```

```
dfinal1 = dfinal.drop(columns=['Cód_Municipio','Nome_Municipio'])
```

```
dfinal1
```

```
import folium
from folium.plugins import HeatMap
```

```
map = folium.Map(location=[-22.8, -50.7], zoom_start=7)
total_data = dfinal1[['latitude','longitude',' ISDEL ']].groupby(['latitude','longitude']).sum().reset_index().values.tolist()
grad = {'0.2': 'Navy','0.25': 'Blue', '0.3': 'Green', '0.35': 'Yellow','0.45': 'Red','0.6':'Purple'}

HeatMap(total_data, radius=8, gradient=grad, max_zoom=13).add_to(map)

map
```
