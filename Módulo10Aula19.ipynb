```
import pandas as pd
import requests

headers = {
    'Content-Type': 'application/json;charset=UTF-8',
    'User-Agent': 'google-colab',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
}
```

```
districts = requests.get("https://servicodados.ibge.gov.br/api/v1/localidades/estados/35/municipios", headers=headers).json()

nome_sp = [municipios["nome"] for municipios in districts]
id_sp   = [municipios["id"] for municipios in districts]
data_sp = pd.DataFrame.from_dict({'ID':id_sp,'MUNICIPIOS':nome_sp})
data_sp
```

```
url = 'https://raw.githubusercontent.com/julianovale/pythonparatodos/main/ISDEL_Cidades_Estado_SPIBGE.csv'
metrica = pd.read_csv(url, encoding='latin-1',sep=";", decimal=',')
metrica.dropna(inplace=True)
metrica
```

```
result = pd.merge(left=data_sp,right=metrica,left_on='ID',right_on='Cód_Municipio')
result
```

```
final_data = result.drop(columns=['Cód_Municipio','Nome_Municipio'])
final_data
```
