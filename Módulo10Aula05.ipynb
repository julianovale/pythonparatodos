```
import pandas as pd
```

```
url = 'https://raw.githubusercontent.com/kelvins/Municipios-Brasileiros/main/csv/municipios.csv'
df = pd.read_csv(url)
df
```

```
print(df.columns.ravel())
```

```
sp = df.loc[df['codigo_uf']==35]
sp
```

```
df3 = df.loc[df['codigo_uf']==35, ['nome', 'latitude','longitude']]
df3
```

```
df3.reset_index(inplace=True)
df3.drop(['index'], axis=1, inplace=True)
df3
```

```
import folium
```

```
map = folium.Map(location=[-22.7864889,-50.6786708], zoom_start = 7) # parâmetros iniciais de visualização do mapa
map
```

```
for i in range(0,len(df3)):
  pop = str(df3.iloc[i]['nome'])
  pop = pop.encode('latin1')
  folium.Marker([df3.iloc[i]['latitude'],df3.iloc[i]['longitude']],popup=pop).add_to(map)

map
```
