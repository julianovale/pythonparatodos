```
import pandas as pd
import io
import requests
```

```
url = 'https://raw.githubusercontent.com/julianovale/pythonparatodos/main/total_cases.csv'
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
df
```

```
df.plot()
```

```
dft = df.copy()
dft = dft.transpose()
dft
```

```
dft = dft.dropna()
dft
```

```
import matplotlib.pyplot as plt
x = df['date']
y1 = df['China']
y2 = df['China']

plt.plot(x,y1,label = 'China')
plt.plot(x,y2,label = 'China2')
plt.legend()
plt.title("Evolução de casos confirmados")
plt.show()
```
